```python
import pandas as pd
```

```python
df = pd.read_csv("Sales.csv")
df
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
# Convert 'Order Date' to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True, errors='coerce')
df['Order Date']
```

```python
# Drop rows where Order Date couldn't be parsed
df = df.dropna(subset=['Order Date'])
df
```

```python
import warnings
warnings.filterwarnings("ignore")
# Step 4: Create a new column for Month-Year
df['Month-Year'] = df['Order Date'].dt.to_period('M').astype(str)
df['Month-Year']
```

```python
# Check for any missing or negative Sales/Profit values
df = df[(df['Sales'] >= 0) & (df['Profit'].notnull())]
df
```

```python
# Save cleaned data for Tableau
df.to_csv("Cleaned_Sales.csv", index=False)
```
